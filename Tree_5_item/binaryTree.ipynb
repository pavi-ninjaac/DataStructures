```python
from QueueLinkedList import Queue
```

```python
class Tree:
    def __init__(self,data):
        self.value = data
        self.left = None
        self.right = None

        
        
def preorder(root):
    if root is None:
        return
    print(root.value)
    preorder(root.left)
    preorder(root.right)

def inorder(root):
    if root is None:
        return
    
    preorder(root.left)
    print(root.value)
    preorder(root.right)

def postorder(root):
    if root is None:
        return
    
    postorder(root.left)
    postorder(root.right)
    print(root.value)

    
def levelorder(root):
    if root is None:
        return 
    else:
        customqueue = Queue()
        customqueue.enqueue(root)
        while not(customqueue.isEmpty()):
            root =customqueue.dequeue()
            print(root.value.value)
            if root.value.left is not None:
                customqueue.enqueue(root.value.left)
            if root.value.right is not None:
                customqueue.enqueue(root.value.right)
            
    
    
def SearchBT(root,Nodevalue):
    if root is None:
        print("TheTree does not exsists!!!")
    else:
        customqueue = Queue()
        customqueue.enqueue(root)
        while not(customqueue.isEmpty()):
            root =customqueue.dequeue()
            if root.value.value == Nodevalue:
                return "The value present in the Tree"
            if root.value.left is not None:
                customqueue.enqueue(root.value.left)
            if root.value.right is not None:
                customqueue.enqueue(root.value.right)
        return "The value not present in the Tree"

            
def insert(root, newnode):
    if root is None:
        root = newnode
    else:
        customqueue = Queue()
        customqueue.enqueue(root)
        while not(customqueue.isEmpty()):
            rootnode = customqueue.dequeue()
            if rootnode.value.left is not None:
                customqueue.enqueue(rootnode.value.left)
            else:
                rootnode.value.left = newnode
                return " The Node inserted Successfully as left child!!!"
            if rootnode.value.right is not None:
                customqueue.enqueue(rootnode.value.right)
            else:
                rootnode.value.right = newnode
                return " The Node inserted Successfully as right child!!!"
            
            
    
def getDeebestNode(root):
    if root is None:
        return 
    else:
        customqueue = Queue()
        customqueue.enqueue(root)
        while not(customqueue.isEmpty()):
            root =customqueue.dequeue()
            
            if root.value.left is not None:
                customqueue.enqueue(root.value.left)
            if root.value.right is not None:
                customqueue.enqueue(root.value.right)
        deebestNode = root.value
        return deebestNode
def deleteDeebestNode(root, dNode):
    if root is None:
        return 
    else:
        customqueue = Queue()
        customqueue.enqueue(root)
        while not(customqueue.isEmpty()):
            root =customqueue.dequeue()
            if root.value is dNode:
                root.value = None
                return 
            if root.value.left is not None:
                if root.value.left is dNode:
                    root.value.left=None
                    return
                else:
                    
                    customqueue.enqueue(root.value.left)
            if root.value.right is not None:
                if root.value.right is dNode:
                    root.value.right=None
                    return
                else:
                    customqueue.enqueue(root.value.right)
        deebestNode = root.value
def deleteNode(root,nodevalue):
    if root is None:
        return 
    else:
        customqueue = Queue()
        customqueue.enqueue(root)
        while not(customqueue.isEmpty()):
            root = customqueue.dequeue()
            if root.value.value == nodevalue:
                deepestNode = getDeebestNode(root.value)
                root.value.value = deepestNode.value
                deleteDeebestNode(root.value,deepestNode)
                return "Node deleted Successfully!!!"
            if root.value.left is not None:
                customqueue.enqueue(root.value.left)
            if root.value.right is not None:
                customqueue.enqueue(root.value.right)
    return "Node deletion failed!!!!"
```
